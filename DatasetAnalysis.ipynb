#### Analysis of the Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.load('data/datasets/dataset_for_analysis.npz')

fs=data['fs']
zz=data['zz_matrices']
es=data['energies']
n_sites=zz.shape[-1]
```

```python
plt.hist(fs,bins=30)
plt.show()
```

```python
print(es.shape)
```

```python
plt.figure(figsize=(10,10))
for i in range(es.shape[1]):
    plt.hist(es[:,i],bins=30,label=f'state={i}')
plt.legend(fontsize=30)
plt.show()
    
```

```python
de=es[:,-1]-es[:,0]

plt.hist(de/n_sites,bins=30)
plt.show()

for i in range(es.shape[0]):
    
    if de[i]/n_sites <0.02:
        print(es[i],'\n')
        plt.imshow(zz[i])
        plt.colorbar()
        plt.show()
```
