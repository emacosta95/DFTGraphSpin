#### Analysis of the Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.load('data/datasets/dataset_for_training_ndata_10000_nsites_16_test0.npz')

fs=data['fs']
zz=data['zz_matrices']
density_f=data['fs_density']
es=data['energies']
n_sites=zz.shape[-1]
```

```python
plt.hist(fs,bins=30)
plt.show()
```

```python
print(es.shape)
```

```python
plt.figure(figsize=(10,10))
for i in range(es.shape[1]):
    plt.hist(es[:,i],bins=60,label=f'state={i}',alpha=0.6)
plt.legend(fontsize=30)
plt.show()

# Sample color palette (you can expand this if you have more states)
colors = plt.cm.tab20(np.linspace(0, 1, es.shape[1]))
plt.figure(figsize=(12, 8))

for i in range(es.shape[1]):
    plt.hist(
        es[:, i],
        bins=60,
        label=f'state={i}',
        alpha=0.5,
        color=colors[i],
        edgecolor='black',
        histtype='stepfilled'  # or 'step' for outline-only
    )

plt.legend(fontsize=20)
plt.xlabel(r'$e_i$',fontsize=30)
plt.ylabel('Frequency',fontsize=30)
#plt.title('Histogram of States',fontsize=30)
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.tick_params(which='major',labelsize=30)
plt.show()
```

```python
de=es[:,-1]-es[:,0]

plt.hist(de/n_sites,bins=30)
plt.show()

plt.figure(figsize=(8, 6))

plt.hist(
    de / n_sites,
    bins=30,
    color='steelblue',          # Choose a color or match from earlier palette
    edgecolor='black',
    alpha=0.6,
    histtype='stepfilled'       # Same style as before
)

plt.xlabel('e / n_sites',fontsize=30)
plt.ylabel('Frequency',fontsize=30)
#plt.title('Histogram of e / n_sites')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.tick_params(which='major',labelsize=30)
plt.show()

```
